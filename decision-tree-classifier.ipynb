# **Decision Trees - Sklearn - Bikes**

```python
#Importando o pandas
import pandas as pd
```

```python
# importando a tabela de treino
treino_x=pd.read_csv('/kaggle/input/edciedadosuniceub2022/df_train.csv')
treino_x.head()
```

# Tratando os dados

### Renomeando as colunas e excluindo a coluna data

```python
rename={'ride_date':'data',
        'tavg': 'temp_media',
        'tmin':'temp_min',
        'tmax':'temp_max',
        'wspd':'vento',
        'pres':'pressao',
        'dia_da_semana':'dia_semana',
        'classe':'bom'}
treino_x=treino_x.rename(columns=rename)
treino_x.drop(columns=['data'],inplace=True)
treino_x.head()
```

### Separando a tabela de treino y(coluna chamada 'bom') da tabela de treino X

```python
treino_y=pd.DataFrame(treino_x.bom)
treino_x.drop(columns=['bom'],inplace=True)
treino_y.head()
```

### Importando os dados de teste e renomeando as colunas


```python
teste_x=pd.read_csv('/kaggle/input/edciedadosuniceub2022/df_submission.csv')
#criando dicionario para renomear
rename={'ride_date':'data',
        'tavg': 'temp_media',
        'tmin':'temp_min',
        'tmax':'temp_max',
        'wspd':'vento',
        'pres':'pressao',
        'dia_da_semana':'dia_semana',
        'classe':'bom'}
teste_x=teste_x.rename(columns=rename)
#tirando a coluna data
teste_x.drop(columns=['data'],inplace=True)

teste_x.head()
```

### Importando bibliotecas matematicas e de machine learning

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

# Árvore de decisão padrão

## Inicializando a arvore de decisao e treinando o modelo 

```python
modelo=DecisionTreeClassifier(max_depth=4)
modelo.fit(treino_x,treino_y)
previsoes=modelo.predict(teste_x)
previsoes_col=previsoes.tolist()
print(previsoes_col)
```

# Random forest

### Importando os dataframes novamente para a implementação dos algoritmos RANDOM FOREST

```python
# importando a tabela de treino
treino_x=pd.read_csv('/kaggle/input/edciedadosuniceub2022/df_train.csv')
#renomeando as colunas
rename={'ride_date':'data',
        'tavg': 'temp_media',
        'tmin':'temp_min',
        'tmax':'temp_max',
        'wspd':'vento',
        'pres':'pressao',
        'dia_da_semana':'dia_semana',
        'classe':'bom'}
treino_x=treino_x.rename(columns=rename)
#tirando a coluna data e copiando ela para outro dataframe
treino_x.drop(columns=['data'],inplace=True)
treino_y=pd.DataFrame(treino_x.bom)
treino_x.drop(columns=['bom'],inplace=True)
#importando o teste x e renomeando
teste_x=pd.read_csv('/kaggle/input/edciedadosuniceub2022/df_submission.csv')
rename={'ride_date':'data',
        'tavg': 'temp_media',
        'tmin':'temp_min',
        'tmax':'temp_max',
        'wspd':'vento',
        'pres':'pressao',
        'dia_da_semana':'dia_semana'}
teste_x=teste_x.rename(columns=rename)
teste_x.drop(columns=['data'],inplace=True)
```

## Criando funções facilitadoras

```python
%%time
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

#treina o modelo e da o predict
def executar_classificador(classificador, X_train, X_test, y_train):
  arvore = classificador.fit(X_train, y_train)
  y_pred = arvore.predict(X_test)
  return y_pred


#criando função que salva a arvore em png
def salvar_arvore(classificador, nome):
  plt.figure(figsize=(200,100))
  tree.plot_tree(classificador, filled=True, fontsize=14)
  plt.savefig(nome)
  plt.close()
    
#criando função que tira as metricas de acuracia, precisão, recall e matrix de confusão
def validar_arvore(y_test, y_pred):
  print(accuracy_score(y_test, y_pred))
  print(precision_score(y_test, y_pred))
  print(recall_score(y_test, y_pred))
  print(confusion_matrix(y_test, y_pred))
```

# Random Forest (BAGGING)

```python
%%time
from sklearn.ensemble import RandomForestClassifier

#inicializando o algoritmo random forest bagging

classificador_random_forest = RandomForestClassifier(n_estimators=80, random_state=0, max_depth=10)
y_pred_random_forest = executar_classificador(classificador_random_forest, treino_x, teste_x, treino_y.values.ravel())
y_pred_random_forest=y_pred_random_forest.tolist()
print(y_pred_random_forest)
```

# Random forest (BOOSTING)

```python
%%time
#importando o algoritmo adaboost
from sklearn.ensemble import AdaBoostClassifier

classificador_adaboost = AdaBoostClassifier(random_state=0, n_estimators=100)
y_pred_adaboost = executar_classificador(classificador_adaboost, treino_x, teste_x, treino_y.values.ravel())
y_pred_adaboost=y_pred_adaboost.tolist()
print(y_pred_adaboost)
```

# Train Test Split

```python

#lendo os dados de treino novamente
dados=pd.read_csv('/kaggle/input/edciedadosuniceub2022/df_train.csv')
dados.drop(columns=['ride_date'],inplace=True)
rename={'tavg': 'temp_media',
        'tmin':'temp_min',
        'tmax':'temp_max',
        'wspd':'vento',
        'pres':'pressao',
        'dia_da_semana':'dia_semana',
        'classe':'bom'}
dados=dados.rename(columns=rename)
dados.head()






```

```python
#biblioteca de divisao de treino de teste
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier


#separando o que é x e o que é y
X = dados[['temp_media', 'temp_min', 'temp_max', 'vento', 'pressao', 'dia_semana']]
y = dados["bom"]

#separando em treino e teste 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
#fazendo o modelo fazer a previsão
#arvore simples de decisão
modelo=DecisionTreeClassifier(max_depth=10)
#treinando o modelo
modelo.fit(X_train,y_train)
#dando predict
previsoes=modelo.predict(X_test)
```

# Metricas avaliativas

```python
#biblioteca de precisão
from sklearn.metrics import precision_score
#biblioteca de acuracia
from sklearn.metrics import accuracy_score
#biblioteca de matriz de confusão
from sklearn.metrics import confusion_matrix

#mostrando a precisao
print('A precisão foi de : ',precision_score(y_test, previsoes))
#mostrando a acuracia
print('A acuracia foi de : ',accuracy_score(y_test, previsoes))
#mostrando a matriz de confusao
print('Matriz de confusão:\n',confusion_matrix(y_test, previsoes))
```
